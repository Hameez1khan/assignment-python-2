```python
import pandas as pd
```

```python
data = pd.read_csv("Pokemon.csv")
```

```python
data
```

```python
data["Speed"].mean()
```

```python
Speed_mean = int(data["Speed"].mean())
def set_speed(val):
    if val < Speed_mean :
       return "Low Speed"
    elif val == Speed_mean :
       return "Mean Speed"
    else : 
       return "High Speed"
```

```python
data["High_Low_Speed"] = data["Speed"].apply(set_speed)
```

```python
data
```
